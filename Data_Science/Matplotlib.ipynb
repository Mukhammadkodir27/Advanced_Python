```python
plt.plot([2, 4, 6, 8], [4, 8, 12, 16], color='green')
plt.xlabel('X coordinates', fontsize=16, color='blue')
plt.ylabel('Y coordinates', fontsize=16, color='red')
```

```python
x = np.linspace(start=0, stop=10, num=50)
plt.plot(x, np.sin(x), label='Sine Curve')
plt.legend() # by default it will take label for legend value
plt.xlabel('X Coordinates', fontsize=16, color='green')
plt.ylabel('Y Coordinates', fontsize=16, color='green')
plt.tick_params(axis='y', color='red', labelcolor='purple' ,labelsize=13)
plt.tick_params(axis='x', color='red', labelcolor='purple', labelsize=13)
plt.xlim(1, 5)
plt.ylim(-1, 0.5)
```

# Lines & Markers

```python
plt.plot(x, np.sin(x), label='Sine Curve', color='green')
plt.plot(x, np.cos(x), label='Cosine Curve', color='m')
plt.legend()
plt.title('Trigonometric Plot')
```

```python
x = np.linspace(start=0, stop=10, num=50)
fig = plt.figure(figsize=(5, 3))
ax1 = fig.add_axes([0, 0.6, 1, 0.4])
ax2 = fig.add_axes([0, 0, 0.8, 0.4])

ax1.plot(x, np.sin(x))
ax1.set_xlabel('X', fontsize=15, color='r')
ax1.set_ylabel('Cos(Y)', fontsize=15, color='r')
ax2.plot(x, np.cos(x))
ax2.set_xlabel('X', fontsize=15, color='r')
ax2.set_ylabel('Cos(X)', fontsize=15, color='r')
```

```python
fig = plt.figure(figsize=(5, 3))
ax1 = fig.add_axes([0, 0, 1, 1])
ax2 = fig.add_axes([0.5, 0.5, 0.4, 0.4])
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(221)
ax1.plot([0, 1.5, 3, 4], [6.7, 6, 7, 5])
ax2 = fig.add_subplot(222)
ax2.plot(x, np.sin(x))
ax3 = fig.add_subplot(223)
ax3.plot(x, np.cos(x))
```

```python
ax1 = plt.subplot2grid((2, 3), (0,0))
ax1.plot(x, np.sin(x))
ax1.set_label("Sine Curve")

ax2 = plt.subplot2grid((2, 3), (0,1))
ax2.plot(x, np.cos(x))
ax2.set_label("Cos Curve")

ax3 = plt.subplot2grid((2, 3), (0,2), rowspan=2)
ax3.plot([1, 2, 3, 4], [2, 4, 6, 8])
ax3.set_label('Straight Line')

ax4 = plt.subplot2grid((2, 3), (1,0), colspan=2)
ax4.plot(x, np.exp2(x))
ax4.set_label('Bottom Plot Line')
```
